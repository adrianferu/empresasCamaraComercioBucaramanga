# Paso 5: Merge entre los dataframes geocidificados

Habiendo culminado con el proceso de geocodificaci'on, podemos proceder a hacer un merge entre los 4 dataframe sobtenidos:



## 5.1. HAcemos el merge

```python
#Creamos los dataframes a partir de los archivos .csv que generamos:
geocod_2020 = pd.read_csv('CH_camComercioExtra_2020_Georref.csv')
geocod_2021 = pd.read_csv('camComercioExtra_2021_Georref_Google.csv')
geocod_2022 = pd.read_csv('CH_camComercioExtra_2022_Georref.csv')
geocod_2023 = pd.read_csv('camComercioExtra_2023_Georref.csv')

# Lista de DataFrames
dataframes = [geocod_2020,geocod_2021,geocod_2022,geocod_2023]
# Fusionar los DataFrames
camComecioGeocoded = pd.concat(dataframes, ignore_index=True)
```

```python
camComecioGeocoded.tail(10)
```

```python
#ELiminamos la columna unnamed:0
#camComecioGeocoded = camComecioGeocoded.drop(columns=['Unnamed: 0'], inplace=True)
```

## 5.2. Hacemos merge con el dataframe original:

```python
# Realizar el merge
result_dataframe = pd.merge(empresas_ambNuevas, camComecioGeocoded[['coords','DIREC_HERE']], how='inner', on='DIREC_HERE')

# Mostrar el nuevo DataFrame resultante
result_dataframe.tail(10)
```

```python
#Eliminamos duplicados: sabemos que enr ealidad son 27.792 empresas
result_dataframe  = result_dataframe.drop_duplicates(subset='RAZON_SOCIAL', keep='first')
result_dataframe.reset_index(inplace = True)
result_dataframe.drop(columns = 'index',inplace = True)
result_dataframe.tail(10)
```

## 5.3. Hacemos la separaci'on de la columna 'coords' en sus coordenadas

```python
#Antes debemos filtrar aquellos valores que nos dieron: ('NA', 'NA') y Error
valores_a_eliminar= ["('NA', 'NA')", "Error"]
# Filtrar las filas que no coincidan con los valores a eliminar
result_dataframe = result_dataframe[~result_dataframe['coords'].isin(valores_a_eliminar)]
result_dataframe.reset_index(inplace = True)
result_dataframe.drop(columns = 'index',inplace = True)

len(result_dataframe)
```

```python
# Expresión regular para extraer latitud y longitud
result_dataframe['latitud'] = result_dataframe['coords'].str.extract(r'\((.*?),')[0].astype(float)
result_dataframe['longitud'] = result_dataframe['coords'].str.extract(r',\s*(.*?)\)')[0].astype(float)
result_dataframe.tail(5)
```

## 5.5. COnvertimos el dataframe a geodataframe

```python
#Ahora convertimos esta dataframe a un Geodataframe
result_Geodataframe = gpd.GeoDataFrame(
    result_dataframe, geometry=gpd.points_from_xy(result_dataframe.longitud, result_dataframe.latitud), crs="EPSG:4326"
)
```

```python
result_Geodataframe.tail(5)
```

```python
result_Geodataframe.to_csv("camaraComercio2020_2023Georref.csv")
```
