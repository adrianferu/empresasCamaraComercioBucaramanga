Verificamos si existen registros duplicados. Para este caso, se usa la columna NIT, debido a que hace referencia al número de identificación de la empresa

## Nota:

Detectamos que hay 14 registros duplicados. Procedemos a eliminarlos

```python
empresas_uniques = empresas.drop_duplicates(keep='first',subset=['NIT','RAZON_SOCIAL'])
empresas_uniques.reset_index(inplace=True,drop=True)#Reiniciamos index para que no hayan errores de numeracion
print(f""" De {len(empresas)} que habian originalmente, ahora hay {len(empresas_uniques)}""")
```

Para este paso-posiblemente el más importante- vamos a usar la API de Geocoding de here, para transformar los datos de dirección a coordenadas geográficas: LATITUD Y LONGITUD. Los pasos para hacer esto son los siguientes:

Recordemos que la columna DIRECCION_COMERCIAL es la que tiene la informaci'on de donde reside la sede de la empresa.
Ejemplo: reemplazar # por No.

2. HAcemos la concatenaci'on entre la columna DIRECCION_COMERCIAL -que ya deber'ia estar limpia- junto con el nombre del municipio (i.e columna CIUDAD) y el nombre del pa'is. Se hace esto con fines de optimizar la b'usqueda

```python
empresas_ambNuevas['DIRECCION_COMERCIAL'] = empresas_ambNuevas['DIRECCION_COMERCIAL'].apply(lambda x : x.replace('#', 'No') )
```

def get_coord (address, YOUR_API_KEY):
    url = f'https://geocode.search.hereapi.com/v1/geocode?q={address}&apiKey={YOUR_API_KEY}'
    
    try:
        response = requests.get(url).json() #Transformacion de la peticion a Json
        dirLimpia = response['items'][0]['title'].upper()#Indexing del json para extraer la direccion
        lat = response['items'][0]['position']['lat'] #Indexing del json para extraer la latitud
        lng = response['items'][0]['position']['lng'] #Indexing del json para extraer la longitud
        
        results = [dirLimpia, lat, lng] #Creamos una lista a partir de los valores d elas 3 variables
        
    except:
        results = ['Not Found', 'NA','NA']
        
    return results
